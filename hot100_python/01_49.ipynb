```python
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        words = {}
        for i,word in enumerate(strs):
            w =  str(sorted(word))
            if w in words:
                words[w].append(word)
            else:
                words[w] = [word]
        return list(words.values())
        


```
