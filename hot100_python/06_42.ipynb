```python
class Solution:
    def trap(self, height: List[int]) -> int:
        n = len(height)
        if n<3:
            return 0 

        left = 0 
        right = n-1
        left_highest = height[left]
        right_highest = height[right]
        water = 0
        while left < right-1:

            if left_highest <= right_highest:
                left += 1
            if left_highest >= right_highest:
                right -= 1
            
            if left == right:
                water += max(min(left_highest,right_highest)-height[left],0)
                break
            water += max(min(left_highest,right_highest)-height[left],0) + max(min(left_highest,right_highest)-height[right],0)
            
            left_highest = max(left_highest,height[left])
            right_highest = max(right_highest,height[right])


       

        return water
```
