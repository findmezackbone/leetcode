```python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        left = 0
        right = len(height)-1
        water_max = 0
        while right>left:
            water = (right-left) * min(height[left],height[right])
            water_max = water if water>water_max else water_max
            if height[left] <= height[right]:
                left += 1
            else:
                right-=1
        
        return water_max

```
