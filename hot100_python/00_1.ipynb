```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        dict_ = {}
        for i in range(len(nums)):
            dict_[target - nums[i]] = i
        
        for i in range(len(nums)):
            if nums[i] in dict_ and i != dict_[nums[i]]:
                return [i,dict_[nums[i]]]

            
```
