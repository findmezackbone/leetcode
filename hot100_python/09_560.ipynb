```python
class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
      
        ans = 0
        pre = 0
        pre_dict = {pre:1}
        for i,num in enumerate(nums):
            pre += num
            ans += pre_dict.get(pre - k,0)
            if pre not in pre_dict:
                pre_dict[pre] = 1
            else:
                pre_dict[pre] += 1
        return ans
```
