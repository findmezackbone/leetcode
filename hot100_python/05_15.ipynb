```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        
        solution = []
        n = len(nums)
        nums = sorted(nums)
        for p1 in range(n):
            if p1>0 and nums[p1] == nums[p1-1]:
                continue
            p2 = n-1
            target = -nums[p1]
            for p3 in range(p1+1,n):
                if p3>p1+1 and nums[p3] == nums[p3-1]:
                    continue
                while p3<p2 and nums[p2]+nums[p3]> target:
                    p2-=1
                if p2 == p3:
                    break
                if nums[p2] + nums[p3] == target:
                    solution.append([nums[p1],nums[p2],nums[p3]])
        
        return solution
                
```
