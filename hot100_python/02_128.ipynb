```python
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        list_start = []
        set_nums = set(nums)
        if nums == []:
            return 0
        for num in nums:
            if num-1 not in set_nums and num not in list_start:
                list_start.append(num)
        i = 1
        j = 1
        for num in list_start:
            i=1
            while num+i in set_nums:
                i+=1
            j = i if i>j else j    
        return j 
```
