```python
from typing import Counter, List
class Solution:
    def findAnagrams(self, s: str, p: str) -> List[int]:
        print(Counter(p))
        n = len(s)  # 字符串s的长度
        p_len = len(p)  # 字符串p的长度
        p_word = {}  # 用于存储字符串p中字符及其出现次数的字典
        ans = []  # 存储结果的列表，存放所有异位词的起始索引

        # 统计字符串p中每个字符出现的次数
        for word in p:
            if word in p_word:
                p_word[word] += 1
            else:
                p_word[word] = 1
        left = 0
        right = 0
        s_word = p_word.copy()
        while left<n and right<n:
            if s[left] not in p_word:
                left += 1
                right = left
                s_word = p_word.copy()
            else:
                if s[right] not in s_word:
                    left = right
                    
                    continue
                if s_word.get(s[right],0) > 0:
                    s_word[s[right]] -= 1
                    right += 1
                else:
                    s_word[s[left]] += 1
                    left += 1
   
            if (right - left ) == p_len:
                if all(value == 0 for value in s_word.values()):
                    ans.append(left)
                
        return ans
    
if __name__ == "__main__":
    s = "cbaebabacd"
    p = "abc"
    print(Solution().findAnagrams(s,p))
    s1 = "abacbabc"
    print(Solution().findAnagrams(s1,p))
```
