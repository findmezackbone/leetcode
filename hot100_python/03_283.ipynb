```python
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        
        moving_zeros = 1 if nums[0]==0 else 0
        for i in range(1,len(nums)):
            if nums[i] == 0:
                moving_zeros+=1
            if nums[i] != 0 and moving_zeros > 0:
                nums[i-moving_zeros]=nums[i]
                nums[i] = 0
            
```
