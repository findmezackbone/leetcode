```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        slow = 0
        fast = 1
        n = len(s)
        max_length = 1
        length = 1
        dict_ = {s[0]:0} if s else None
        while slow<n and fast<n :
            max_length = max(max_length,length)
            if s[fast] in dict_:
               
                
                slow = dict_[s[fast]]+1
                dict_ = {s[slow]:slow}
                fast = slow + 1
                length = 1
            else:
                dict_[s[fast]] = fast
                length += 1
                fast+=1
        max_length = max(max_length,length)    
        return min(n,max_length)

             
```
